# Loading Image Data 

I had previously created candle stick charts for some SP500 stocks. I will load a subset of them using `datasets.ImageFolder` from `torchvision`. In general Imagefolder is used like: 

```python 
dataset = datasets.ImageFolder('path/to/data', tranform = transforms)
``` 

where `'path/to/data'` is the file path to data directory and transforms is a list of processing steps built with `transforms` modeul with `torchvision`. 
The data directory we feed into ImageFolder should be as follows:

    root/class1/xx1.png
    root/class1/xx2.png
    root/class1/xx3.png
    ...
    root/class2/yy1.png
    root/class2/yy2.png
    root/class2/yy3.png


````python
import torch
from torchvision import datasets, transforms
````

# Model Transfering 

We will use one the models trained on ImageNet data for our problem here. For the sake of example we will use DenseNet model. Most of the built models require the input to be 224x224 images. Each color channel is normalized separately, the means are `[0.485, 0.456, 0.406]` and the standard deviations are `[0.229, 0.224, 0.225]`. 

````python
from torchvision import models 
````

````python
transformation = transforms.Compose([ ])
````

````python
model = models.densenet121(pretrained = True) 
model
````

We have features and classifier parts. We will not touch the features part, but change the classifier. Originally the last part of the model was `(classifier): Linear(in_features=1024, out_features=1000, bias=True)`. So there are 1024 input parameters and 1000 output features. However, we have only 3 outputs. So we will turn off the gradients for the features and train the classifier part of the model. 

````python
from torch import nn 

for param in model.parameters():
    param.requires_grad = False 

from collections import OrderedDict
classifier = nn.Sequential(OrderedDict([
    ('fc1', nn.Linear(1024, 500)),
    ('relu', nn.ReLU()),
    ('fc2', nn.Linear(500, 3)),
    ('output', nn.LogSoftmax(dim=1))
]))

model.classifier = classifier 
model.cuda()
````
